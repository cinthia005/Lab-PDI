**Parte I.- Procesamiento geométrico**

Ejercicio 1: Traslación
Se desplazaran los píxeles de una imagen en dirección horizontal (x) y vertical (y) con una matriz de transformación.

```python
#Importar Librerias
import cv2 #procesar imagen
import numpy as np #operaciones y matrices
import matplotlib.pyplot as plt #visualizacion
```

```python
#Mostrar imágenes en escala de grises
def mostrar_img(titulo, img):
    plt.imshow(img, cmap='gray')
    plt.title(titulo)
    plt.axis('off')
    plt.show()
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
# Cargar imagen en escala de grises
img = cv2.imread('Radiografia.jpg', cv2.IMREAD_GRAYSCALE)

# Traslación entera (50, 30)
M = np.float32([[1, 0, 50], [0, 1, 30]])
traslacion1 = cv2.warpAffine(img, M, (img.shape[1], img.shape[0]))

# Traslación decimal (20.5, 15.5)
M_decimal = np.float32([[1, 0, 20.5], [0, 1, 15.5]])
traslacion2 = cv2.warpAffine(img, M_decimal, (img.shape[1], img.shape[0]))

# Mostrar resultados
mostrar_img("Imagen Original", img)
mostrar_img("Traslación (50,30)", traslacion1)
mostrar_img("Traslación (20.5,15.5)", traslacion2)
```

Ejercicio 2. Rotación

```python
(h, w) = img.shape
centro = (w//2, h//2)

# Rotación de 45 grados
M_rot = cv2.getRotationMatrix2D(centro, 45, 1.0)
rotada = cv2.warpAffine(img, M_rot, (w, h))

# Mostrar resultados
mostrar_img("Imagen Original", img)
mostrar_img("Rotación 45°", rotada)
```

Ejercicio 3. Escala

```python
# Escala 150%
escala_up = cv2.resize(img, None, fx=1.5, fy=1.5, interpolation=cv2.INTER_LINEAR)

# Escala 50%
escala_down = cv2.resize(img, None, fx=0.5, fy=0.5, interpolation=cv2.INTER_LINEAR)

mostrar_img("Imagen Original", img)
mostrar_img("Escala 150%", escala_up)
mostrar_img("Escala 50%", escala_down)
```

### **Parte II – Ecualización de histograma**
Ejercicio 1: Ecualización


```python
from google.colab import files
uploaded = files.upload()
```

```python
# Cargar imagen de bajo contraste
img2 = cv2.imread('RM2.jpg', cv2.IMREAD_GRAYSCALE)  # segunda imagen médica

# Histograma original
hist_original = cv2.calcHist([img2], [0], None, [256], [0,256])

# Ecualización
img_eq = cv2.equalizeHist(img2)
hist_eq = cv2.calcHist([img_eq], [0], None, [256], [0,256])

# Mostrar imágenes
mostrar_img("Imagen Original (Bajo contraste)", img2)
mostrar_img("Imagen Ecualizada", img_eq)

# Mostrar histogramas
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.plot(hist_original, color='blue')
plt.title("Histograma Original")
plt.subplot(1,2,2)
plt.plot(hist_eq, color='red')
plt.title("Histograma Ecualizado")
plt.show()
```
